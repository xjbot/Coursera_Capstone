##  Reading the Postcode and name data

```python
previous_form=pd.read_csv('Organized_geo.csv')

previous_form.drop(['Unnamed: 0'],axis=1, inplace=True)
previous_form
```

## Reading the coordinate data

```python
geo_data=pd.read_csv('Geospatial_Coordinates.csv')

geo_data 
```

## change the name of the shared column

```python
geo_data1=geo_data.rename(columns={'Postal Code':'PostCode'} )     
geo_data1
 
```

## merging the two data frame into one

```python
full_data=pd.merge( previous_form, geo_data1,   on=['PostCode'])

full_data
```
